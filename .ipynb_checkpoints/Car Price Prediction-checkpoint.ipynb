### Feature Engineering

```python
# Adding a new feature - "number of year" of the car used
```

```python
df.columns
```

```python
# Getting rid of Car name as it will not help our model with prediction. Also, we need all numerical data

final_dataset = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataset.head()
```

```python
final_dataset['Current_Year'] = 2022
```

```python
final_dataset['Num_of_Year'] = final_dataset['Current_Year'] - final_dataset['Year']
```

```python
final_dataset.head()
```

```python
# deleting the Year and Current_Year column

final_dataset.drop(['Year', 'Current_Year'], axis=1, inplace=True)
```

```python
final_dataset.head()
```

```python
# Convert categorical variables to numerical values using get_dummies()

final_dataset = pd.get_dummies(final_dataset, drop_first=True)
```

```python
final_dataset.head()
```

```python
# Finding Correlation

final_dataset.corr()
```

```python
# Pairplot to see visually how features are correlated

sns.pairplot(final_dataset)
```

```python
# X and y seperation of features

X = final_dataset.iloc[:,1:]
y = final_dataset.iloc[:,0] # contains the Selling_Price
```

```python
X.head()
```

```python
y.head()
```

### Performing feature importance

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
# Lets see the list of features with higher importance value

temp = pd.DataFrame(model.feature_importances_, index=X.columns)
temp
```

```python
feat_imp = pd.Series(model.feature_importances_, index=X.columns)
feat_imp.sort_values(ascending = False).plot(kind = 'barh')
plt.show()
```

These are the top important features 

### Modelling

```python
# performing train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
# Check the performance of random forest regressor and then perform hyper parameter tuning. 
# HyperParameterTuning is taking multiple value for various parameters and then checking which value yields best results

# This is also required to avoid overfitting
```

```python
rf.fit(X_train, y_train)
```

```python
y_predictions = rf.predict(X_test)
```

```python
sns.distplot(y_test-y_predictions)
```

```python
sns.scatterplot(x=y_test, y=y_predictions)
plt.xlabel('Y Test Values')
plt.ylabel('Y Predicted Values')
plt.show()
```

```python
# Calculate RMSE and also r2 score

from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("RMSE is {}".format(np.sqrt(mean_squared_error(y_test, y_predictions))))
```

```python
print("R2_score is {}".format(r2_score(y_test, y_predictions)))
```

```python
# NOW perform HyperParameter Tuning using RamdomisedSearchCV to see if we yield better score

from sklearn.model_selection import RandomizedSearchCV
```

```python
# Setting up different parameters

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
param_dict = {'n_estimators':n_estimators,
             'max_features':max_features,
             'max_depth':max_depth,
             'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_dict, )
```
